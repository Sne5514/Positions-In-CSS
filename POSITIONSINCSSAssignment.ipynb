# **Task 2:**


**Problem Statement**

Explain the difference between Absolute and Relative positioning.

 -  

***1. Absolute Positioning (position: absolute;)***

-The element is removed from the normal document flow.

-It is positioned relative to the nearest positioned ancestor (i.e., an ancestor with position: relative;, absolute;, or fixed;).

-If no such ancestor exists, it is positioned relative to the <html> (the entire page).

-The space it originally occupied is no longer reserved.

-Example:

    .absolute-box {
    position: absolute;
    top: 20px;
    left: 20px;
    }

-This moves the element 20px from the top-left of its nearest positioned ancestor (or the page if none exist).

-Positioned relative to the closest positioned ancestor or the document body if no ancestor is found.

-Moves along with the page when scrolling.

-Does not retain its original space; other elements may occupy it.

-Used for exact positioning within a container or page.

-Creates a new stacking context based on the ancestorâ€™s stacking order.

-Precise positioning within a specific container or overlay effects.

**Example:**It is an example of position: absolute; property. The absolute element is positioned 80px from the top and right of its closest positioned ancestor (the relative container).

```
<!DOCTYPE html>
<html>
<head>
  <style>
        div.relative {
            position: relative;
            width: 400px;
            height: 200px;
            border: 3px solid #73AD21;
        }
        div.absolute {
            position: absolute;
            top: 80px;
            right: 80px;
            width: 200px;
            height: 100px;
            border: 3px solid #73AD21;
        }
    </style>
</head>
<body>
    <h1>position: absolute;</h1>
    <div class="relative">
        This element has position: relative;
        <div class="absolute">
            This element has position: absolute;
        </div>
    </div>
</body>
</html>
```

***2. Relative Positioning (position: relative;)***

-The element is positioned relative to its normal position in the document flow.

-Using top, left, right, or bottom moves the element from where it would normally be without affecting other elements.

-The space it originally occupied remains reserved.

-Example:

    .relative-box {
    position: relative;
    top: 20px;
    left: 20px;
    }

-This moves the element 20px down and 20px right from its original location.

-Positioned relative to its normal location in the flow.

-Moves along with the page when scrolling.

-Retains its original space in the document layout.

-Used to move an element slightly from its original position.

-Can create a new stacking context if z-index is applied.

-Minor adjustments like nudging elements.

**Example:** It is an example of position: relative; property. The element is shifted 50px to the right using the left property, but it still occupies its original space in the document flow.

```
<!DOCTYPE html>
<html>

<head>
      <style>
        div.relative {
            position: relative;
            left: 50px;
            border: 3px solid #73AD21;
        }
    </style>
</head>

<body>
     <h1>position: relative;</h1>
     <div class="relative">
     This element has position:relative;
    </div>
 </body>

</html>
```

# **Task 5:**


**Problem Statement**

Explain the z-index, with a code example.

The z-index property in CSS controls the vertical stacking order of elements. Elements with a higher z-index value appear in front of those with a lower value. It only works on elements with a position of relative, absolute, or fixed.

**Example:**



```
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Z-Index Example</title>
  <style>
        .container {
            position: relative;
            width: 300px;
            height: 300px;
            border: 2px solid black;
        }
        .box {
            position: absolute;
            width: 100px;
            height: 100px;
            color: white;
            text-align: center;
            line-height: 100px;
            font-weight: bold;
        }
        .box1 {
            background-color: red;
            top: 50px;
            left: 50px;
            z-index: 1;
        }

        .box2 {
            background-color: blue;
            top: 70px;
            left: 70px;
            z-index: 2;
        }
        .box3 {
            background-color: green;
            top: 90px;
            left: 90px;
            z-index: 3;
        }
    </style>
</head>
<body>
    <div class="container">
        <div class="box box1">Z-1</div>
        <div class="box box2">Z-2</div>
        <div class="box box3">Z-3</div>
    </div>
</body>
</html>


```
